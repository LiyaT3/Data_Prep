```python
df.isna().sum()
```

```python
import seaborn as sns
import numpy as np
sns.boxplot(x=df['Planned_TimeofTravel'])
```

```python
#FOR A TRIAL VERSION,DOING ALL 4 METHODS SO CREATING 4 COPIES TO SHOW EACH METHOD

trial_1=df.copy()
trial_2=df.copy()
trial_3=df.copy()
trial_4=df.copy()
trial_5=df.copy()
```

```python
#METHOD 1-REMOVAL OF OUTLIERS

iqr1=trial_1['Planned_TimeofTravel'].quantile(0.75)-trial_1['Planned_TimeofTravel'].quantile(0.25)
lower_lt=trial_1['Planned_TimeofTravel'].quantile(0.25)-(iqr1*1.5)
upper_lt=trial_1['Planned_TimeofTravel'].quantile(0.75)+(iqr1*1.5)
```

```python
#TRYING TO REMOVE THE OUTLIERS AS THE PROBABILITY OF THE SAME PATTERN HAPPENING IS LESS

outliers_df=np.where(trial_1['Planned_TimeofTravel']>upper_lt,True,np.where(trial_1['Planned_TimeofTravel']<lower_lt,True,False))
trial_1=trial_1.loc[~outliers_df]
```

```python
trial_1.shape
```

#still outliers present,going for method 2

```python
#METHOD 2 -RETAIN

iqr2=trial_2['Planned_TimeofTravel'].quantile(0.75)-trial_2['Planned_TimeofTravel'].quantile(0.25)
lower_lt2=trial_2['Planned_TimeofTravel'].quantile(0.25)-(iqr2*1.5)
upper_lt2=trial_2['Planned_TimeofTravel'].quantile(0.75)+(iqr2*1.5)

trial_2['Planned_TimeofTravel']=np.where(trial_2['Planned_TimeofTravel']>upper_lt2,upper_lt2,np.where(trial_2['Planned_TimeofTravel']<lower_lt2,lower_lt2,trial_2['Planned_TimeofTravel']))
```
